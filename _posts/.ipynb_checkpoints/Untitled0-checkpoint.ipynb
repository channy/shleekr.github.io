```python
import fst
t = fst.Transducer()
t.add_arc(0,1, 'R', 'R')
t.add_arc(1,2, 'o', 'o')
t.add_arc(2,3,'ε', 'u')
t.add_arc(3,4,'s','z')
t.add_arc(4,5,'s','ε')
t.add_arc(5,6,'e','e')
t.add_arc(6,7,'t','t')
t.add_arc(7,8,'t','ε')
t.add_arc(8,9,'a','a')
t[9].final=True
```
